```python
import pandas as pd

def calculate_accuracy(file_path):
    try:
        # 读取 CSV 文件
        df = pd.read_csv(file_path)

        # 检查是否存在 prediction 和 label 列
        if 'prediction' not in df.columns or 'label' not in df.columns:
            print("文件中缺少 'prediction' 或 'label' 列，请检查文件。")
            return

        # 计算混淆矩阵的四个指标
        TP = sum((df['prediction'] == 1) & (df['label'] == 1))  # True Positives
        FP = sum((df['prediction'] == 1) & (df['label'] == 0))  # False Positives
        TN = sum((df['prediction'] == 0) & (df['label'] == 0))  # True Negatives
        FN = sum((df['prediction'] == 0) & (df['label'] == 1))  # False Negatives

        # 计算准确率
        accuracy = (TP + TN) / (TP + FP + TN + FN) * 100  # 百分比形式

        # 输出结果
        print(f"True Positives (TP): {TP}")
        print(f"False Positives (FP): {FP}")
        print(f"True Negatives (TN): {TN}")
        print(f"False Negatives (FN): {FN}")
        print(f"预测准确率为：{accuracy:.2f}%")

    except FileNotFoundError:
        print(f"未找到文件：{file_path}，请检查文件路径是否正确。")
    except Exception as e:
        print(f"发生未知错误：{e}")

if __name__ == "__main__":
    # 请将下面的文件路径替换为你实际的文件路径
    file_path = 'A0201_100000_test.csv'
    calculate_accuracy(file_path)
```
